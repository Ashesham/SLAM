```python
import cv2
import numpy as np

def is_corner(pixels, center_pixel, threshold):
    darker_threshold = center_pixel - threshold
    brighter_threshold = center_pixel + threshold
    
    dark_pixels = np.sum(pixels < darker_threshold)
    bright_pixels = np.sum(pixels > brighter_threshold)
    
    return dark_pixels >= 9 or bright_pixels >= 9

def detect_FAST_keypoints(image, threshold=100):
    image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    keypoints = []
    
    for y in range(3, image.shape[0] - 3):
        for x in range(3, image.shape[1] - 3):
            pixels = image_gray[y - 2:y + 3, x - 2:x + 3].ravel()
            center_pixel = image_gray[y, x]
            
            if is_corner(pixels, center_pixel, threshold):
                keypoints.append(cv2.KeyPoint(x, y, 7))
    
    return keypoints

# Load an image
image_path = '../data/imgs/1.png'
image = cv2.imread(image_path)

# Detect FAST keypoints
keypoints = detect_FAST_keypoints(image)

# Draw the keypoints on the image
keypoints_image = cv2.drawKeypoints(image, keypoints, None, color=(0, 0, 255))

# Display the result
cv2.imshow('FAST Keypoints', keypoints_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

def detect_FAST_keypoints(image, threshold=100):
    image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    dy_minus_2 = image_gray[2:-2, 3:] - image_gray[4:, 3:]
    dy_plus_2 = image_gray[4:, 3:] - image_gray[2:-2, 3:]
    dx_minus_2 = image_gray[3:, 2:-2] - image_gray[3:, 4:]
    dx_plus_2 = image_gray[3:, 4:] - image_gray[3:, 2:-2]
    
    corner_mask = (np.abs(dy_minus_2) > threshold) & \
                  (np.abs(dy_plus_2) > threshold) & \
                  (np.abs(dx_minus_2) > threshold) & \
                  (np.abs(dx_plus_2) > threshold)
    
    y_coords, x_coords = np.where(corner_mask)
    y_coords += 3  # Adjust for the cropping
    x_coords += 3  # Adjust for the cropping
    
    keypoints = [cv2.KeyPoint(x, y, 7) for x, y in zip(x_coords, y_coords)]
    
    return keypoints

# Load an image
image_path = '../data/imgs/1.png'
image = cv2.imread(image_path)

# Detect FAST keypoints
keypoints = detect_FAST_keypoints(image)

# Draw the keypoints on the image
keypoints_image = cv2.drawKeypoints(image, keypoints, None, color=(0, 0, 255))

# Display the result
cv2.imshow('FAST Keypoints', keypoints_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
